```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
# now here scaling is done by dividing /255 since the value will be in betwwen 0 and 1
X_train=X_train/255
X_test=X_test/255
```

```python
model = Sequential()

# now since the values is in the form of a matrix 28x28 we need the 784 so we flatten
# convert matrix to array
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob=model.predict(X_test)
# for every image an array of 0 prob,1 prob, 2 prob.....9 prob
```

```python
# now we have probability for all can be nos
# so the max of the prob is the number in this image
y_prob.argmax(axis=1)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
